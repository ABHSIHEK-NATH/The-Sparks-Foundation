Finding pairwise Correlation between Coloumns and Visualize using Heatmap

```python
superstore.corr(numeric_only=True)
plt.figure(figsize=(12,6))
sns.heatmap(superstore.corr(numeric_only=True), annot=True)
plt.show()
```

Univariate Analysis

```python
superstore['Ship Mode'].value_counts()
```

```python
superstore['Ship Mode'].value_counts().plot(kind = 'pie', figsize=[10,10], autopct='%1.1f%%')
plt.title('Shipping mode wise order distribution')
```

_Inference - Standard shipping is highly preferred while same day shipping is least preferred_

```python
superstore['Segment'].value_counts()
```

```python
superstore['Segment'].value_counts().plot(kind= 'pie', figsize=[10,10],autopct='%1.1f%%')
plt.title('Segmentwise distribution')
```

_Inference - We can see that 52% people belong to consumer class whereas 17% to 30% belong to Home office & corporate_

```python
superstore['Country'].value_counts()
```

_Inference - We can see that all the orders are under United States_

```python
superstore['Category'].value_counts()
```

```python
superstore['Category'].value_counts().plot(kind='pie', figsize= [10,10], autopct='%1.1f%%')
plt.title('Categorywise distribution')
```

_Inference - We can see that about 60% category belongs to office supplies whereas approxmately 20% category belongs to furniture and technological supplies_

```python
superstore['Sub-Category'].value_counts()
```

```python
superstore['Sub-Category'].value_counts().plot(kind='pie', figsize=[10,10], autopct='%1.1f%%')
plt.title('Sub-Category wise distribution')
```

_Inference - Binders followed by papers are the most popular sub-category_

```python
superstore['City'].value_counts().head(20)
```

```python
superstore['City'].value_counts().head(20).plot(kind='barh')
plt.title('Top 20 cities having most orders placed')
```

_Inference - Maximum number of orders are placed in New York City_

```python
superstore['State'].value_counts().head()
```

```python
superstore['State'].value_counts().sort_values(ascending= True).plot(kind='bar',figsize= [20,6])
plt.title('State wise order distribution')
```

_Inference - Maximum orders are placed in California_

```python
superstore['Region'].value_counts()
```

```python
superstore['Region'].value_counts().plot(kind='pie', figsize= [10,10], autopct= '%1.1f%%')
plt.title('Regional order distribution')
```

_Inference - Western region has the maximum sales of 32%_

Bivariate Analysis

```python
categorical_sum= superstore.groupby('Category').sum(numeric_only= True)
print(categorical_sum)
```

```python
categorical_sum[['Sales', 'Profit']].plot.bar()
```

_Inference - Technological supplies have highest sales and profit. furnitures have moderate sales but least profit. Office supplies have least sales but moderate profit_

```python
subcategorical_sum= superstore.groupby('Sub-Category').sum(numeric_only= True)
subcategorical_sum
```

```python
subcategorical_sum[['Sales','Profit']].plot(kind= 'bar', figsize= [20,8])
plt.ylabel('Total Sales/Profit', fontsize= 15)
```

_Inference - Here Tables have the lowest profit and Copiers have the highest profit. Also Fasteners have the lowest sales whereas Phones have the highest sales._

```python
superstore.groupby(by='State')['Profit'].sum().sort_values(ascending= True).plot(kind='bar',figsize= [20,8])
plt.ylabel('Profit', fontsize=15)
```

_Inference - Here the Maximum Profit is in California and Maximum Loss is in Texas._

```python
plt.figure(figsize=(20,10))
sns.lineplot(x='Discount', y='Profit', data=superstore)
plt.title('Discount and profit')
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(x='Discount', y='Sales', data=superstore)
plt.title('Discount versus Sales')
```

_Inference - It shows that if we give more discounts on products sales will increase but profits will fall._
